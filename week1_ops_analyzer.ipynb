**Purpose:** This notebook calculates operational efficiency for a set of business sites, classifies each site's status, and prints a clean report. Today it's retail stores, but the same logic works for fuel depots, hospital wards, or any operation that has a target vs actual output.

```python
def calculate_efficiency(actual_output, target_output):
    """
    Calculate efficiency as a percentage of actual output vs target output.
    """
    if target_output == 0:
        raise ValueError("target_output cannot be zero — efficiency is undefined.")

    efficiency = (actual_output / target_output) * 100
    return round(efficiency, 2)


# Quick sanity check
print(calculate_efficiency(80, 100))  # Expect 80.0
print(calculate_efficiency(45, 50))   # Expect 90.0
```

## Step 1: KPI Function — calculate_efficiency

Efficiency is just "how much we actually produced" compared to "how much we were supposed to produce", turned into a percentage.

The formula is: efficiency = (actual_output / target_output) * 100

If a store sold 80 units against a target of 100, that's (80 / 100) * 100 = 80% efficient.

One thing to watch for: never divide by zero. If target_output is 0, the formula breaks, so we guard against that below.

```python
def get_operational_status(efficiency):
    """
    Classify an efficiency percentage into an operational status.
    """
    if efficiency < 70:
        return "Critical"
    elif efficiency < 90:
        return "Warning"
    else:
        return "Normal"


# Quick sanity check
print(get_operational_status(65))   # Expect Critical
print(get_operational_status(80))   # Expect Warning
print(get_operational_status(95))   # Expect Normal
```

## Step 2: Status Classifier — get_operational_status

Once we know the efficiency number, we still need to translate it into something a manager can act on at a glance. This function takes a number and turns it into a plain-English status label.

The rules:
- Below 70% → Critical (something is seriously wrong, needs immediate attention)
- 70% up to 89% → Warning (underperforming, keep an eye on it)
- 90% and above → Normal (operating as expected)

Python checks each condition top to bottom and stops at the first one that's true, so the order matters here.

```python
sites = [
    {"name": "Westgate Branch", "actual_output": 850, "target_output": 1000},
    {"name": "CBD Branch", "actual_output": 620, "target_output": 950},
    {"name": "Karen Branch", "actual_output": 980, "target_output": 1000},
]
```

## Step 3: Data Modeling — 3 Retail Stores

In a real operational system, you wouldn't track one site — you'd track many. The cleanest way to represent "many similar things, each with the same set of properties" is a list of dictionaries.

Each entry represents one store, and every store has the same three fields: name, actual_output, and target_output. This mirrors what you'd get back from a database query or an API call in a real job — a list of records, each with consistent fields.

Chosen industry: 3 Retail Stores. actual_output and target_output represent monthly sales units.

```python
for site in sites:
    name = site["name"]
    actual = site["actual_output"]
    target = site["target_output"]

    efficiency = calculate_efficiency(actual, target)
    status = get_operational_status(efficiency)

    print(f"Site: {name}")
    print(f"  Actual Output: {actual}")
    print(f"  Target Output: {target}")
    print(f"  Efficiency: {efficiency}%")
    print(f"  Status: {status}")
    print("-" * 40)
```

## Step 4: Execution — Loop, Calculate, Classify, Report

Now we bring it all together. This is the "Operations → Data → Insight → Action" loop in code form:

1. Operations already happened — that's where actual_output came from.
2. Data is sitting in our sites list.
3. We loop through each site, since we have a fixed list to step through.
4. For each one, we call calculate_efficiency() to turn the raw numbers into a percentage — that's the Insight.
5. We pass that into get_operational_status() to get a status label that's easy to act on.
6. We print a formatted report — this is what would actually go in front of a manager to drive the Action.
